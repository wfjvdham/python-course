# Type of Joins

![title](img/join-image.jpg)

```python
import pandas as pd

flights = pd.read_csv("../datasets/flights.csv")
airlines = pd.read_csv("../datasets/airlines.csv")
airports = pd.read_csv("../datasets/airports.csv")
planes = pd.read_csv("../datasets/planes.csv")
weather_nyc = pd.read_csv("../datasets/weather_nyc.csv")
```

Left join without specifying the columns

```python
flights.merge(airlines, how='left')[['dep_delay', 'carrier', 'name']]
```

Left join on multiple columns without specifying them

```python
flights.merge(weather_nyc, how='left')[['dep_delay', 'hour', 'temp']]
```

Left join when the same column name is present in both datasets but the join is not on this column name

```python
flights.merge(planes, how='left', on='tailnum')[['year_x', 'year_y', 'engine']]
```

Join on different column names

```python
flights.merge(airports, how='left', left_on='dest', right_on='faa')[['dest', 'alt']]
```
