# Seaborn

Using `set()` the style of plots created is changed to look a bit more modern.

```python
import seaborn as sns
sns.set()

titanic_df = sns.load_dataset("titanic")
titanic_df
```

## One Variable

| Type `x` | *First* Graphic |
| --- | --- |
| Discrete | **countplot** |
| Continuous | **histogram** |

```python
sns.distplot(titanic_df['fare'])
```

```python
sns.countplot(x="sex", data=titanic_df)
```

## Two Variables

| Type of `x` | Type of `y` | *First* Graphic |
| --- | --- | --- |
| Discrete | Discrete | **countplot** |
| Discrete | Continuous | **boxplot** |
| Continuous | Continuous | **lmplot** |

**Note** when there are a lot of data points the last plot will be slow. You can use `cut()` to divide one variable in groups and use a boxplot after that.

```python
sns.countplot(x="sex", hue="survived", data=titanic_df)
```

```python
fig = sns.boxplot(x="survived", y="fare", data=titanic_df)
fig.set(ylim=(-10, 200))
```

```python
sns.lmplot(x="age", y="fare", data=titanic_df)
```

## Three Variables

| Type of `x` | Type of `y` | Type of `z` | *First* Graphic |
| --- | --- | --- | --- |
| Discrete | Continuous | Discrete | **barplot** with *hue* |
| Continuous | Continuous | Discrete | **lmplot** with *color* |
| Continuous | Continuous | Continuous | **scatterplot** with *hue* |

```python
sns.barplot(x="sex", y="age", hue="pclass", data=titanic_df)
```

```python
sns.lmplot(x="fare", y="age", hue="pclass", data=titanic_df)
```

```python
sns.scatterplot(x="fare", y="parch", hue="age", data=titanic_df)
```

## Title and Labels

```python
fig = sns.barplot(x="sex", y="age", hue="pclass", data=titanic_df)
fig.axes.set_title("Test title", fontsize=16)
fig.set_xlabel("Gender", fontsize=14)
fig.set_ylabel("Age", fontsize=14)
```
